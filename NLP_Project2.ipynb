#Import Library

```python
import soundfile as sf
import torch
from transformers import Wav2Vec2Processor, Wav2Vec2ForCTC
import time
import os
import librosa
import re
from pytube import YouTube
from moviepy.editor import *
import math
import datetime
```

# Load Pretrained Model

```python
# load pretrained processor & model 
# Wav2Vec train 53 ภาษา finetune เป็นไทย
processor = Wav2Vec2Processor.from_pretrained("airesearch/wav2vec2-large-xlsr-53-th")
model = Wav2Vec2ForCTC.from_pretrained("airesearch/wav2vec2-large-xlsr-53-th")
```

# Natural Sorting (Sorting sound file by alphabet) 

```python
def atoi(text):
    return int(text) if text.isdigit() else text

def natural_keys(text):
    '''
    alist.sort(key=natural_keys) sorts in human order
    http://nedbatchelder.com/blog/200712/human_sorting.html
    (See Toothy's implementation in the comments)
    '''
    return [ atoi(c) for c in re.split(r'(\d+)', text) ]
```

# Download Video from Youtube

```python

link = "https://www.youtube.com/watch?v=JsNjj8t1e8o"
yt = YouTube(link)
yt = yt.streams.filter(progressive=True, file_extension='mp4').order_by('resolution').desc().first()
yt.download("/content/input")
```

# Generate Sound files from Video

```python
#Configuration 

index=0
input_folder = "/content/input/"
root_output_folder = "/content/output/"

if os.path.isdir(input_folder) is False:
  os.mkdir(input_folder)

if os.path.isdir(root_output_folder) is False:
  os.mkdir(root_output_folder)
```

```python

for root, dirs, files in os.walk(input_folder, topdown=False):
   for name in files:
      print(name)
      
      output_folder=os.path.splitext(root_output_folder+name)[0]+"/"
      print(output_folder)
      audio,sr = librosa.load(input_folder+name)
      clips = librosa.effects.split(audio, top_db=40)
      #print(clips)
      index=0
      if os.path.isdir(output_folder) is False:
        os.mkdir(output_folder)
      for c in clips:
        print(c)
        data = audio[c[0]: c[1]]
        sf.write(output_folder+"sound"+str(index)+'.wav', data, sr)
        index = index+1

```

# Use a model to predict speech


```python
num = 0
total_length = 0
start = time.time()

for root, dirs, files in os.walk(root_output_folder, topdown=False):
    files.sort(key=natural_keys)
    for name in files:
        # load audio
        if name[-4:] == ".txt" or name[-4:] == ".TXT":
            continue
        num += 1
        dirname = root
        print("no. : ",num)
        print("name : "+name)
        print("File Path : "+dirname)
        if(name=="sound0.wav"):
          f = open(dirname+".txt", "w")
        else:
          f = open(dirname+".txt", "a")
        audio_input, fs = sf.read(os.path.join(root, name))
        total_length += audio_input.shape[0] / fs
        audio_input_16 = librosa.resample(audio_input, fs, 16000)
        input_values = processor(audio_input_16,return_tensors="pt", sampling_rate=16000).input_values
        print(input_values.shape)

        logits = model(input_values).logits
        predicted_ids = torch.argmax(logits, dim=-1)
        transcription = processor.batch_decode(predicted_ids)[0]
        print(transcription+"\n")
        f.write(transcription+"\n")
        f.close()


print("time: ", time.time() - start)
print("total speech length: ", total_length)
print("total files: ", num)
```
